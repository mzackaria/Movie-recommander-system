# Movie recommendation system study - Part 2

## Introduction

In a previous part we studied a movie dataset from the imdb databases. We've pointed out what was the most important data for the task at hand: building a recommendation system. The data is preprocessed and ready and we are now trying to figure out the best item to item model to fit our data - ie approximatively 5k movies, and build the best recommendation system.

Our goal is to make a recommendation from one movie to 5 others. We're going to use three main strategies to build up our recommendation system: 
 - A recommendation based on a distance feature
 - A recommendation based on a hierarchical clustering
 - A recommendation based on a K-mean clustering


## 1 - Distances

 The first we can do is very simple: sent back the 5 closest movies in the dimensional space that we created (which we think represent the optimal way to differentiate movies). 


```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('movie_preprocessed.csv', index_col=0)
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(data)
```

We're going to try out different distances to use for our recommendation system. 

To select the most performing distance there is no better way to look than directly to try it out and look for movie's recommendations for each distance. Than we'll assess which one we "like" the best.

```python
from sklearn.neighbors import DistanceMetric
```

### Manhattan distance

```python
dist_manhattan = DistanceMetric.get_metric('manhattan')
manhattan = dist_manhattan.pairwise(X)
```

### Euclidian distance 

```python
dist = DistanceMetric.get_metric('euclidean')
euclidian = dist.pairwise(X)
```

### Chebyshev Distance 

```python
dist = DistanceMetric.get_metric('chebyshev')
chebyshev = dist.pairwise(X)
```

### Minkowski Distance  

```python
dist = DistanceMetric.get_metric('minkowski', p = 10)
minkowski = dist.pairwise(X)
```

##### To compare the distances let's just look at the closest neighbor for the first 5 films for each distance.

```python
for i in range(len(euclidian)):
    euclidian[i,i] = 100
    manhattan[i,i] = 100
    chebyshev[i,i] = 100
    minkowski[i,i] = 100

#nearest movie name for each distance
euclidian = data.index.values[np.argmin(euclidian, axis = 0)].astype(str)
manhattan = data.index.values[np.argmin(manhattan, axis = 0)].astype(str)
chebyshev = data.index.values[np.argmin(chebyshev, axis = 0)].astype(str)
minkowski = data.index.values[np.argmin(minkowski, axis = 0)].astype(str)

#dataframe 
values = np.concatenate([euclidian.reshape((4288,1)), manhattan.reshape((4288,1)), chebyshev.reshape((4288,1)), minkowski.reshape((4288,1))], axis = 1)
index = data.index.values
columns = ['euclidian', 'manhattan', 'chebyshev', 'minkowski']
result = pd.DataFrame(values, index = index, columns = columns)
```

```python
result.head(20)
```

#### Conclusion on distances

First we can notice that this simple technique is really effective: if I enter a James Bond Movie it returns an other James Bond movie as a recommendation (line 3).

Now we want to know which one of those distances is the best one. What we do is nothing magical: we just compare the recommendation for each distance and each movie.

We can notice that it seems like euclidian and manhattan distances tend to give better results than the chebyshev and minkowski distances (purely subjective observation).

If I had to choose one distance, I would personnaly choose the euclidian distance for faster computing of the distances and what seems to be more accurate results.

Let's have a look at the different results we would have with the euclidian distance:

```python
dist = DistanceMetric.get_metric('euclidean')
euclidian = dist.pairwise(X)
```

```python
argument = np.argsort(euclidian)[:,1:6]
pd.DataFrame(argument, index=data.index).applymap(lambda x : data.index[x]).head(20)
```

The first thing that comes out of those results is that we have pretty accurate results for a method that didn't require a lot of computing. 

The problem with this method is that for the same movie it would return the same five recommendations over and over. That's why we switch to "real" machine learning and the two algorithms we're going to use: hierarchical clustering and k-means algorithm.

## Hierarchical clustering

A hierarchical clustering is also an algorithm based on distances. For that we are going to use a dendogram: a graph which is separating our data into different clusters based on distances. 

Two ways we can build cluster on this graph: 
 - bottom up (agglomerative): we try to build clusters by merging data
 - top down (divisive): we try to build clusters by separating data
 
 The question is now how do we decide at what point we merge or divide the data. Well, it depends on what criterium we want to use. For example we can use:
 

<center> - max{ d(a,b) : $a \in A$, $b \in B$ }</center>
<center> OR</center>
<center> - min{ d(a,b) : $a \in A$, $b \in B$ } </center>


The one we're going to use is called **Ward method**. It basically computes the total variance for each cluster and try to find combinations that minimize this criterium as best as it can.

As you can imagine this requires a lot of computation.

```python
# Using the dendrogram to find the optimal number of clusters
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Films')
plt.ylabel('Euclidean distance')
plt.show()
```

It is quite difficult to find a clear cut in this graph - ie a cut that can make clusters that have approximatively the same number of elements.

We can do a number of things to find out if we can still use this method:
- Change the criterium
- change the distance which is used to compute the criterium
- change the dataset: although we are going to put aside information tha might be usefull for clustering, taking a subset of our dataset might give us better cuts and improve computation time.

Just to give an example of what it might look like, we only took from our dataset the genre features and applied the previous method:

```python
%matplotlib inline
genres = 0.3*data.iloc[:,65:92]
dendrogram = sch.dendrogram(sch.linkage(genres, method = 'ward'))
plt.title('Dendrogram on genre features')
plt.xlabel('Films')
plt.ylabel('Euclidean distances')
plt.plot(range(0, 500000), 17*np.ones(500000))
plt.show()
```

We can see that it is easier to find a cut in the graph that might give us  good clusters but as we've said we've put a side a LOT of information so the clusters aren't going to be really relevant.

This method hasn't shown any significant results so we're going to use a much more efficient algorithm: K-means

# K-means

We are going to use a very famous algorithm: K-means. This algorithm is very effective to build clusters without prior knowledge of your data.

The only issue with that algorithm is that you first have to determine K - ie the number of clusters. To do that you must use different K values and see which is the best one. To determine which is the best K, a good practice is to use the elbow method (more on that later).

This means that we're going to run this algorithm a few time. 

This algorithm is quite heavy (computing time is  rather long) so we're going to use a PCA to reduce the dimensionality of our data first.

#### PCA

```python
from sklearn import decomposition

pca = decomposition.PCA(n_components=0.8)
pca.fit(X)
X_PCA = pd.DataFrame(pca.transform(X), index = data.index)
```

#### Silhouette value

Now what we want is to determine K the number of cluster. We have 2 ways to look at this. First we can concentrate on the silhouette value. The silhouette value is a measure of how similar an object is to its own cluster.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

score = []
wcss = []

for i in range(5, 150, 5):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(X_PCA)
    score.append(silhouette_score(X_PCA.values, kmeans.predict(X_PCA)))
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(5, 150, 5), score)
plt.title('Silhouette Score vs number clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

We can notice that the optimal number of clusters looking at that metric is approximatively **110 clusters.**

```python
max(np.array(score))
```

#### Elbow method

The other interesting method is called the **"elbow method"**. It considered the metric WCSS (Within-Cluster-Sum-of-Squares), which represents the variance within a cluster.


<center> WCSS = $\sum_{}^{}$  ($X_{i}$ - $Y_{i}$ )$^{2} $  </center>

This value needs to be lowered to a minimum that is interesting. Between the actual minimum (each film is one cluster) and the maximum (every movie within a cluster). The most interesting value can be seen the WCSS(K) graph that we will print below.

```python
plt.plot(range(5, 150, 5), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

The approximative optimal number of cluster with this method is 70 clusters

### Improvement of the model

The question that arises is: can we improve that model ? 

Well, the matter of fact is that YES WE CAN!

First let's recap. We had some data regarding 5k films. We built the features that we thought were interesting to clusterise those movies. And now we are trying out different machine learning models to see which one is the best fit (welcome to a data scientist's life).

What we need to consider is: is there a better way to shape our data so that our K-mean algorithm can better clusterize the data.

A way to better shape the data would be to give more or less "importance" to certain features depending on whether we want to build our clustering on those features or not.

In our previous tries, we decided that all the features had the same importance: meaning the feature "imdb score" had the same weight as the feature "number of facebook likes for the third actor". Now, maybe it would be more interseting to give more weight to the "imdb score" feature than the "number of facebook likes for the third actor" feature as we certainly want our recommendation system to base it's decision more on the "imdb score" feature than the "number of facebook likes for the third actor".

What if we decide to rank the features based on the weight we want to give them in the clustering process.


| Suggestion of ranks | Features name | weight given |
| --------------------|:-------------------:|-------------:|
|1                    |genres               |     1         |
|2                    |score_imdb           |     1         |
|3                    |content-rating       |      0.3        |
|4                    |country              |      0.25        |
|5                    |duration             |      0.2        |
|6                    |director_name        |      0.15        |
|7                    |title_year           |      0.1        |

|10                   |color                |       0.05       |
|11                   |movie_facebook_likes |       0.05         |
|12                   |cast_total_facebook_likes  |  0.05              |





#### Building the dataset to fead the model - weighed features

```python

#function to resize and wheigh a numerical column in the dataset
def resize_numerical_data(name_column, dataframe=data, weight=1, sc=StandardScaler()):
    column_values = dataframe[[name_column]].values.reshape((len(dataframe), 1))
    column_values_resized = weight*sc.fit_transform(column_values)
    return pd.DataFrame(column_values_resized, columns=[name_column], index=dataframe.index)

#function to resize and wheigh a categorical columns with the prefix 'prefix_name_column' in the dataset
def resize_categorical_data(prefix_name_column, dataframe=data, weight=1):
    return weight*dataframe.filter(items=[col for col in dataframe.columns if prefix_name_column in col])

#RESIZING NUMERICAL DATA
imdb_score = resize_numerical_data('imdb_score')
duration = resize_numerical_data('duration', weight=0.2)
title_year = resize_numerical_data('title_year', weight=0.1)
budget = resize_numerical_data('budget', weight=0.1)
movie_facebook_likes = resize_numerical_data('movie_facebook_likes', weight=0.1)
cast_total_facebook_likes = resize_numerical_data('cast_total_facebook_likes', weight=0.1)

#RESIZING CATEGORICAL DATA
genres = resize_categorical_data('genre', weight=0.1)
content_rating = resize_categorical_data('content_rating', weight=0.8)
country = resize_categorical_data('country', weight=0.25)
director_name = resize_categorical_data('director_name', weight=0.15)
actors = resize_categorical_data('actor', weight=0.15)

#BUILDING THE DATASET FOR kMEANS
df = pd.concat([imdb_score, genres, content_rating, country, duration, director_name, title_year, actors, budget, movie_facebook_likes, cast_total_facebook_likes], axis=1)
```

```python
df.head()
```

```python
#ELBOW METHOD
wcss = []
score = []
for i in range(5, 150, 5):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df.values)
    wcss.append(kmeans.inertia_)
    score.append(silhouette_score(df.values, kmeans.predict(df)))
```

```python
plt.plot(range(5, 150, 5), wcss)
plt.title('ELBOW METHOD')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

We can set K now and K is approximatively equals to **40** (according to the graph below).

We can look at 2 or 3 classes just to look at how those classes are built and if our recommendation is satisfying enough.

```python
kmeans = KMeans(n_clusters = 40, init = 'k-means++', random_state = 42)
kmeans.fit(df.values)
y = kmeans.predict(df)
```
